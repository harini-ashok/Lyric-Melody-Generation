```python
import os
import re
import pprint as pp
lmd_artist_track_map={}
for artist in os.listdir("./LMD"):
    if(os.path.isdir("./LMD/"+artist)):
        for track in os.listdir("./LMD/"+artist):
            formatted_artist= " ".join(artist.split("_")).lower()
            formatted_artist= re.sub('[^a-zA-Z0-9\s]', '', formatted_artist).strip()
            
            formatted_track=track.split(".mid")[0]
            formatted_track = " ".join(formatted_track.split("_")).lower()
            formatted_track= re.sub('[^a-zA-Z\s]', '', formatted_track).strip()
            
            if(formatted_artist not in lmd_artist_track_map):
                lmd_artist_track_map[formatted_artist]={}   
            if(formatted_track not in lmd_artist_track_map[formatted_artist]):
                lmd_artist_track_map[formatted_artist][formatted_track]=[]
                
            lmd_artist_track_map[formatted_artist][formatted_track].append(artist+"/"+track)


        
pp.pprint(lmd_artist_track_map)
```

```python
def is_similar(a,b):
    a_parts=set(a.split(" "))
    b_parts=set(b.split(" "))
    if(len(b_parts)==1):
        return a==b
    score=0
    for b_word in b_parts:
        if b_word in a_parts:
            score+=1
    score=score/len(b_parts)
    return score>0.8;
def get_artist_map(artists_lmd,artists_muse):
    matched_count=0
    match_map={}
    for artist_lmd in artists_lmd:
        matched=False
        for artist_muse in artists_muse:
            if(is_similar(artist_muse,artist_lmd)):
                if(artist_lmd not in match_map):
                    match_map[artist_lmd]=[artist_muse]
                else:
                    match_map[artist_lmd].append(artist_muse)
                matched=True
                matched_count+=1
    print("Matched ",matched_count,"/",len(artists_lmd))
    return match_map
        
artist_map=get_artist_map(lmd_artist_track_map,muse_artist_track_map)
```

```python
results=[]
for lmd_artist in artist_map:
    for muse_artist in artist_map[lmd_artist]:
        lmd_tracks=list(lmd_artist_track_map[lmd_artist].keys())
        muse_tracks=list(muse_artist_track_map[muse_artist].keys())
        for lmd_track in lmd_tracks:
            for muse_track in muse_tracks:
                if(is_similar(lmd_track,muse_track)):
                    for filename in lmd_artist_track_map[lmd_artist][lmd_track]:
                        result_line=[filename]
                        result_line.extend(muse_artist_track_map[muse_artist][muse_track])
                        results.append(result_line)
                
print(len(results))                    
pp.pprint(results)
```

```python
import csv
with open('matched.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow(["filepath",'track', 'artist', 'seeds', 'number_of_emotion_tags',
       'valence_tags', 'arousal_tags', 'dominance_tags'])
    writer.writerows(results)
```
